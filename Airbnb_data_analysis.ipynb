```python
# Replace missing values in specific columns
listining_data['name'].fillna("Unknown", inplace=True)
listining_data['host_since'].fillna(listining_data['host_since'].mode()[0], inplace=True)
listining_data['host_location'].fillna("Unknown", inplace=True)
# Continue with similar operations for other columns...

# For numeric columns, you can replace missing values with median or mode
numeric_cols = ['host_total_listings_count', 'bedrooms', 'review_scores_rating', 'review_scores_accuracy', 'review_scores_cleanliness', 'review_scores_checkin', 'review_scores_communication', 'review_scores_location', 'review_scores_value']
for col in numeric_cols:
    listining_data[col].fillna(listining_data[col].median(), inplace=True)
```

```python
# first we need to see the nan values of out data to know how can we deal with it
listining_data.isna().sum()
```

```python
# Handle missing values for remaining columns
listining_data['host_response_time'].fillna("Unknown", inplace=True)
listining_data['host_response_rate'].fillna(listining_data['host_response_rate'].median(), inplace=True)
listining_data['host_acceptance_rate'].fillna(listining_data['host_acceptance_rate'].median(), inplace=True)
listining_data['district'].fillna("Unknown", inplace=True)
```

```python

from scipy.stats import ttest_ind


Super_host_price = listining_data[listining_data['host_is_superhost'] == True]['price']
un_Super_host_price = listining_data[listining_data['host_is_superhost'] == False]['price']


t_statistic, p_value = ttest_ind(Super_host_price, un_Super_host_price)


alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: There is a significant difference in prices between Super_host_price and un_Super_host_price.")
else:
    print("Fail to reject the null hypothesis: There is no significant difference in prices between Super_host_price and un_Super_host_price")
```

```python
average_price_superhost = Super_host_price.mean()
average_price_non_superhost = un_Super_host_price.mean()

# bar plot
plt.figure(figsize=(8, 6))
plt.bar(['Superhost', 'Non-Superhost'], [average_price_superhost, average_price_non_superhost], color=['blue', 'orange'])
plt.xlabel('Host Type')
plt.ylabel('Average Price')
plt.title('Average Price Comparison between Superhosts and Non-Superhosts')
plt.grid(axis='y')
plt.show()
```

```python
# Room Type impact on price
from scipy.stats import f_oneway

# Example room_type
room_type_Entire = listining_data[listining_data['room_type'] == 'Entire place']['price']
room_type_Private = listining_data[listining_data['room_type'] == 'Private room']['price']
room_type_Hotel = listining_data[listining_data['room_type'] == 'Hotel room']['price']
room_type_Shared = listining_data[listining_data['room_type'] == 'Shared room']['price']


f_statistic, p_value = f_oneway(room_type_Entire, room_type_Private, room_type_Hotel , room_type_Shared)


alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: There is a significant difference in prices between different room_type.")
else:
    print("Fail to reject the null hypothesis: There is no significant difference in prices between different room_type.")
```
